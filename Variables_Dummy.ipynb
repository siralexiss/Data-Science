Las variables Dummy son utilizadas para reemplazar datos categoricos, como puede ser datos de una columna de Sexo cuyo valores son Male, Female.

Entonces la variable Dummy se utilizaria para crear dos columnas nuevas, donde serian las columnas de Male , Female y cuyos valores serian 0 o 1 dependiendo si el personaje muestrado es Hombre o Mujer , entiendose como 0 a Falso y como 1 - Verdadero

#Funcion para crear variables Dummy
podemos hacer todo esto pero ya con una funcion, para que sea mas sencillo y ahorarnos pasos, solamente tendremos que ingresar los parametros

```
def createDummies(df,var_name):
	dummy=pd.get_dummies(df[var_name],prefix=var_name)
	df=df.drop([var_name],axis=1)
	df=pd.concat([df,dummy],axis=1)
	return df
```

```
data2=data
data2.head()
```

```
createDummies(data2,'sex')
```
