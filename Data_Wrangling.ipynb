#Metodo Shufle
El metodo de Shuffling sirve para Mezclar los numeros dentro de una lista


```
a=range(100)
b=np.arange(100) #aqui creamos una lista con 100 valores
np.random.shuffle(b)# aqui los mezclamos
b
```
