#Saber que valores son nulos de la columna osea que no tiene valor y esta marcado como NA

```
pd.isnull(data['body'])#Donde utilizamos la funcion es nulo y le ponemos como argumentos nuestro data frame y la columna que queremos analizar, en este caso body
```

```
pd.isnull(data['body']).values.ravel().sum() #para saber la suma de los valores nulos
```

##Borrado de Valores Que faltan

```
data.dropna(axis=0,how='all')
```

Data.dropna se utiliza para eliminar filas o columnas con valores de NaN
si le colocamos el parametro de axis=0 borara filas
y axis=1 borra toda la columna
con el parametro How indicamos con que condiciones la borrara, al ponerle ='all' estamos indicando que solo borre las filas que tengan solo NaN

#Eliminar una fila si almenos una de las columnas tiene Na

```
data2=data
data2.dropna(axis=0,how='any')  #esto elimina la fila si almenos una columna tiene Na y  en este caso se elimino todas las filas, por que almenos una de la columna tenia Na
```

#computo de los valores faltantes, Añadir o reemplazar valores


```
data3=data
data3.fillna(0,inplace=True)#Remplazamos los valores faltantes con 0 y ponemos inplace para que se guarde el cambio
```

```
data4=data
data4=data4.fillna('Desconocido',inplace=True) #remplazar valor faltante con Desconocido o X String
```

#Reemplazar valor de una columna especifica 

```
data5=data
```

```
data5['body'].fillna(0, inplace = True)  #Remmplazar valor de una columna
```

```
data5
```

```
data6=data
```

```
pd.isnull(data6['age']).values.ravel().sum()
```

```
data6['age'].fillna(data['age'].mean(),inplace=True)   #Reemplzar valores faltantes de una columna, con la media de los datos (si la media es 5, entonces los datos faltantes seran 5)
data6
```

#Reemplazar Valores con el Valor inmediato superior o inferior

```
data7=data
data7['age'].fillna(method='ffill',inplace=True) #aqui reemplaza el valor con el anterior inmediato, inplace es para que se guarde el archivo
data7
```

```
data8=data
data8['age'].fillna(method='backfill',inplace=True) #aqui reemplaza el valor con el siguiente inmediato
data8
```
